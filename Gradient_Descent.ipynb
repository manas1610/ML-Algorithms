```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradientdescent(x,y):
    m,b=0,0
    iterations=100
    n=len(x)
    lr=0.01
    plt.scatter(x,y,color='red',marker='+')
    
    for i in range(iterations):
        y_pred=m*x+b
        plt.plot(x,y_pred,color='green')
        cost=(1/(2*n))*sum((val**2)for val in (y-y_pred))
        dm=-(1/n)*sum(x*(y-y_pred))
        db=-(1/n)*sum(y-y_pred)
        m=m-(lr/n)*dm
        b=b-(lr/n)*db
        print(m,b,cost,sep=',')
        
```

```python
x=np.array([1,3,5,9,12])
y=np.array([2,5,8,10,16])
```

```python
gradientdescent(x,y)
```
