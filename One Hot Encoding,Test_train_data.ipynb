```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('F:/Courses/ML & DL Complete/3. Machine Learning/py-master/py-master/ML/5_one_hot_encoding/homeprices.csv')
df.head()
```

```python
dummy=pd.get_dummies(df.town)
dummy
```

```python
df_new=pd.concat([df,dummy],axis='columns')
df_new
```

```python
df_final= df_new.drop(['town','west windsor'],axis='columns')
df_final
```

```python
model=LinearRegression()
x = df_final.drop('price',axis='columns')
x
```

```python
y=df_final['price']
y
```

```python
model.fit(x,y)
```

```python
print('score: ',model.score(x,y))
print()
y_new=model.predict([[3300,1,0]])
y_new
```
